```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = '../model/training_metrics.csv'
data = pd.read_csv(csv_file)
# 读取csv文件
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(data['epoch'], data['loss'],color='blue',label='loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss',color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
# 创建一个X轴维Epoch Y轴为Loss 曲线标签和刻度线为蓝色 方便区分Accuracy

ax1.set_title('Training Loss and Accuracy')
ax1.grid(True)
# 设置图表标题并显示网格线

ax2 = ax1.twinx()
ax2.plot(data['epoch'],data['accuracy'],color='red',label='accuracy')
ax2.set_ylabel('Accuracy',color='red')
ax2.tick_params(axis='y', labelcolor='red')
# 创建另一个Y轴为Accuracy 全部为红色

fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
# 在右上角放置图例

plt.savefig('../model/training_metrics.png')
plt.show()
# 保存并展示
```
